```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib
import matplotlib.pyplot as plt

for_pca = pd.read_csv('for_pca.csv')

topic = for_pca['topic']

data = for_pca.drop(['Unnamed: 0', 'document', 'topic'], axis = 1)
print(data.shape)
ss = StandardScaler().fit(data)

data = ss.transform(data)
print(data.shape)
```

```python
pca = PCA(n_components=2)
pca.fit(data.transpose())
```

```python
comp1 = pca.components_[0]
comp2 = pca.components_[1]

plotting_df = pd.DataFrame({'comp1': comp1, 'comp2': comp2, 'topic': topic})
```

```python
topic_1 = plotting_df.loc[plotting_df['topic'] == 1]
topic_2 = plotting_df.loc[plotting_df['topic'] == 2]
topic_3 = plotting_df.loc[plotting_df['topic'] == 3]
```

```python
plt.figure(figsize=(8,8))
plt.xlim(-.1, .1)
plt.ylim(-.1, .1)
plt.xlabel('First Component', size=15)
plt.ylabel('Second Component', size=15)
plt.title('Topic Distribuation', size = 20)
plt.scatter(topic_3['comp1'], topic_3['comp2'], color='pink', label = 'topic 3')
plt.scatter(topic_2['comp1'], topic_2['comp2'], color='cyan', label = 'topic 2')
plt.scatter(topic_1['comp1'], topic_1['comp2'], color='blue', label = 'topic 1')
#plt.scatter(topic_2['comp1'], topic_2['comp2'], color='cyan', label = 'topic 2')
#plt.scatter(topic_3['comp1'], topic_3['comp2'], color='pink', label = 'topic 3')
plt.legend(loc="lower right")
```

```python
for_pca_age_gen = pd.read_csv('for_pca_age_gender.csv')
gender = for_pca_age_gen['gender_category']
age = for_pca_age_gen['age_category']

data2 = for_pca_age_gen.drop(['Unnamed: 0', 'document', 'gender_category', 'age_category' ], axis = 1)
ss2 = StandardScaler().fit(data2)

#data2 = ss2.transform(data2)

pca2 = PCA(n_components=2)
pca2.fit(data2.transpose())
```

```python
comp1 = pca2.components_[0]
comp2 = pca2.components_[1]

plotting_df2 = pd.DataFrame({'comp1': comp1, 'comp2': comp2, 'age': age, 'gender': gender})
```

```python
men = plotting_df2.loc[plotting_df2['gender'] == 'male']
women = plotting_df2.loc[plotting_df2['gender'] == 'female']
```

```python
plt.figure(figsize=(8,8))
plt.xlim(-.1, .1)
plt.ylim(-.1, .1)
plt.xlabel('First Component', size=15)
plt.ylabel('Second Component', size=15)
plt.title('Gender Distribuation', size = 20)
plt.scatter(men['comp1'], men['comp2'], color='purple', label = 'Men')
plt.scatter(women['comp1'], women['comp2'], color='cyan', label = 'Woman')

plt.legend(loc="lower right")
```

```python

tens = plotting_df2.loc[plotting_df2['age'] == '13-17']
twenties = plotting_df2.loc[plotting_df2['age'] == '23-27']
thirties = plotting_df2.loc[plotting_df2['age'] == '33-48']
```

```python
plt.figure(figsize=(8,8))
plt.xlim(-.1, .1)
plt.ylim(-.1, .1)
plt.xlabel('First Component', size=15)
plt.ylabel('Second Component', size=15)
plt.title('Age Distribuation', size = 20)
plt.scatter(tens['comp1'], tens['comp2'], color='green', label = 'Tens')
plt.scatter(twenties['comp1'], twenties['comp2'], color='orange', label = 'Twenties')
plt.scatter(thirties['comp1'], thirties['comp2'], color='blue', label = 'Thirties')
#plt.scatter(tens['comp1'], tens['comp2'], color='green', label = 'Tens')

plt.legend(loc="lower right")
```
